```python


fig, ax1 = plt.subplots()

ax2 = ax1.twinx()


#It just to have a idea and be able to remember.I have written these type of codes many times but I forget it and this makes it easier for me to recap.
#ax1.plot(x, y, color='blue', label='y') 
ax1.plot(climate_change.index, climate_change["co2"], color="blue", label="CO2")

ax2.plot(climate_change.index, climate_change["relative_temp"], color="red", label="Relative Temp")

ax1.set_xlabel("Time")

ax1.set_ylabel("CO2", color="blue")
ax2.set_ylabel("Relative Temp", color="red")


ax1.set_xlim(pd.to_datetime("2000-01-01"), pd.to_datetime("2016-12-06"))

# Annotate the first date at which co2 exceeded 400
first_date_exceeding_400 = climate_change[climate_change["co2"] > 400].index[0]
ax1.annotate("CO2 > 400", xy=(first_date_exceeding_400, 400), xytext=(first_date_exceeding_400, 420),
             arrowprops=dict(facecolor='black', arrowstyle="->"))

# Annotate the first date at which co2 exceeded 400
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

# Show the plot
plt.show()
```

```python

sports = summer_2016['Sport'].unique()

for sport in sports:
    sport_data = summer_2016[(summer_2016['Sport'] == sport) & (summer_2016['Medal'] == 'Gold')]
    
    plt.hist(sport_data['Age'], bins=20, alpha=0.5, label=sport)

plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age for Gold Medal Winners in Different Sports')

plt.legend()

plt.show()
```

```python
#We will only classify the sports that obtained more gold medals
most_gold_medals = summer_2016[summer_2016['Medal'] == 'Gold'].groupby('Sport')['Medal'].count().sort_values(ascending=False)

print(most_gold_medals.head())
```

```python

most_gold_medals = summer_2016[summer_2016['Medal'] == 'Gold'].groupby('Sport')['Medal'].count().sort_values(ascending=False)

ax = most_gold_medals.plot(kind='barh', figsize=(12, 5), title='Sports with the Most Gold Medals')
ax.set_xlabel('Number of Gold Medals')
ax.set_ylabel('Sport')

plt.show()
```

```python
#Show only the top 10 sports with the most gold medals
#We add the code .nlargest(10) to filter just the biggest 10

most_gold_medals = summer_2016[summer_2016['Medal'] == 'Gold'].groupby('Sport')['Medal'].count().nlargest(10)

ax = most_gold_medals.plot(kind='barh', figsize=(12, 5), title='Top 10 Sports with the Most Gold Medals')
ax.set_xlabel('Number of Gold Medals')
ax.set_ylabel('Sport')

plt.show()

```
