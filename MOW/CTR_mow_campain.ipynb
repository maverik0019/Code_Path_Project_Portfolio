```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
file_path = '/file_path.csv'
final_report_data = pd.read_csv(file_path)
```

```python
group1 = final_report_data[final_report_data['Grupo de anuncios'] == 'Grupo de anuncios 1']
```

```python
group1['Clics'].replace(0, np.nan, inplace=True)  # Reemplazando 0 por NaN para evitar divisiones por cero
```

```python
group1['Conversion Rate'] = group1['Conversion es'] / group1['Clics']
group1['CTR'] = group1['Clics'] / group1['Impr.']
```

```python
conversion_rate_mean = group1['Conversion Rate'].mean()
ctr_mean = group1['CTR'].mean()
```

```python
conversion_rate_mean, ctr_mean
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(group1['Conversion Rate'].dropna(), kde=True, color='blue')
plt.title('Distribucin CR')
plt.xlabel('CR')
plt.ylabel('Frecuencia')


plt.subplot(1, 2, 2)
sns.histplot(group1['CTR'].dropna(), kde=True, color='green')
plt.title('Distribution CTR')
plt.xlabel('CTR')
plt.ylabel('Frecuency')

plt.tight_layout()
plt.show()
```

## Tasa de Conversin (aproximadamente 5.92%):
La tasa de conversin promedio indica que aproximadamente el 5.92% de los clics en los anuncios resultaron en una conversin. Esta es una mtrica crucial, ya que refleja la efectividad de los anuncios para lograr el objetivo final, ya sea una venta, un registro, una descarga, etc.
Dependiendo de la industria y del objetivo especfico del anuncio, esta tasa de conversin puede considerarse alta o baja. En general, una tasa de conversin por encima del 5% es notablemente buena para muchas industrias, pero es importante compararla con las tasas de conversin promedio especficas del sector.

## CTR (aproximadamente 40.37%):
Un CTR de 40.37% es excepcionalmente alto, lo que sugiere que los anuncios son muy relevantes y atractivos para el pblico objetivo. En la mayora de los contextos de publicidad digital, un CTR por encima del 2% ya se considera bueno.
Este alto CTR indica que el diseo del anuncio, el contenido y la seleccin de palabras clave estn bien alineados con las necesidades e intereses de la audiencia objetivo.
